```python
import gradio as gr
```

```python
def correct_sentence(text):
    result = gf.correct(text, max_candidates=1)
    return list(result)[0]
```

```python
spelling_correction_if = gr.Interface(fn = correct_sentence,
                                      inputs="text",
                                      outputs = "text")
```

```python
#Create sharable interface for sentence checking while the cell is running
spelling_correction_if.launch(debug=True)
```
